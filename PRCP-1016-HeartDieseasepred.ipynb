## Insights on Univariate analysis

**Heart Disease Present**

- Binary classification, with 56% of the data indicating "No" (0) and 44% indicating "Yes" (1). This suggests a somewhat balanced target distribution, though slightly skewed toward "No".

**Fasting Blood Sugar > 120 mg/dl**

- 84% of patients have fasting blood sugar levels ≤ 120 mg/dl, while 16% have higher levels, which may indicate a potential risk factor for heart disease.

**Thalassemia Type (Thal)**

- Three categories: "Normal" (54%), "Reversible Defect" (41%), and "Fixed Defect" (4%). Most patients have a normal thalassemia type, with a significant portion showing reversible defects, which could be linked to heart disease.

**Max Heart Rate Achieved**

- The correlation with age is -0.39, suggesting that as age increases, the maximum heart rate tends to decrease, which is expected. There's also a weak correlation with cholesterol and resting blood pressure.

**Slope of Peak Exercise ST Segment**

- Moderate correlation with the target variable (heart disease presence), as well as with chest pain type. The slope is a critical feature in determining heart conditions during exercise.

**Chest Pain Type**
- Different types of chest pain have a strong correlation with heart disease presence, indicating its significance in predicting heart conditions.
